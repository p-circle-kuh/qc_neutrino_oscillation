## Theoretical computation of 2-flavor neutrino oscillation : P($\nu_e \rightarrow \nu_e$)

```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set parameters
# Difference of mass squared between m1 & m2(from NuFIT)
delta_m_12 = 7.42e-5
# Dirac constant(from CODATA)
h_bar = 4.135667696e-15 / (2 * np.pi)
# Theta_12(from NuFIT)
theta_12 = 33.44 / 360 * 2 * np.pi
# Energy of neutrino
Energy = np.arange(2,12,0.05) * 1e6
# Time (L=180[km], c=3e5[km/sec])
t = 180 / 3.0e5
```

```python
def two_flavor_neutrino_oscillation_theory(E):
    return 1.0-0.5*(np.sin(2*theta_12)**2)*(1-np.cos((delta_m_12*t)/(2*E*h_bar)))
```

```python
result = two_flavor_neutrino_oscillation_theory(Energy)
```

```python
# Plot graph
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
ax.plot(Energy,result,label="Theory")
ax.set_xlabel("Energy[Mev]",fontsize=13)
ax.set_ylabel(r"Two Flavor P($\nu$e->$\nu$e)",fontsize=13)
ax.set_xticks([2e6,4e6,6e6,8e6,10e6,12e6])
ax.set_xticklabels(["2","4","6","8","10","12"],fontsize=13)
ax.legend(loc="best",fontsize=13)
ax.tick_params(labelsize=13)
```
